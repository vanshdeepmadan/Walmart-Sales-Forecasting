```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

# Load dataset
file_path = "train_cleaned.csv"  # Update the path if needed
df = pd.read_csv(file_path)

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Aggregate sales at the store level
df_sales = df.groupby("Date")["Weekly_Sales"].sum().reset_index()

# Create only the Lag_52 feature (1-year lag)
df_sales["Lag_52"] = df_sales["Weekly_Sales"].shift(52)

# Drop rows with NaN values (caused by shifting)
df_sales.dropna(inplace=True)

# Define Features (X) and Target (y)
X = df_sales[["Lag_52"]]  # Using only Lag_52
y = df_sales["Weekly_Sales"]

# Split into Train (80%) and Test (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Make Predictions
y_pred = model.predict(X_test)

# Evaluate Model Performance
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"RÂ² Score: {r2:.4f}")

# Plot Actual vs Predicted Sales
plt.figure(figsize=(12, 6))
plt.plot(df_sales["Date"].iloc[len(y_train):], y_test, label="Actual Sales", marker="o")
plt.plot(df_sales["Date"].iloc[len(y_train):], y_pred, label="Predicted Sales", marker="x")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.title("Actual vs Predicted Weekly Sales (Using Lag_52)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import ttest_ind
import pandas as pd

# Load dataset
file_path = "train_cleaned.csv"  # Update if needed
df = pd.read_csv(file_path)

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Aggregate total sales by date
df_sales = df.groupby("Date").agg({"Weekly_Sales": "sum", "IsHoliday_x": "first"}).reset_index()

# Separate sales for holidays and non-holidays
holiday_sales = df_sales[df_sales["IsHoliday_x"] == 1]["Weekly_Sales"]
non_holiday_sales = df_sales[df_sales["IsHoliday_x"] == 0]["Weekly_Sales"]

# Perform independent t-test (Welch's t-test for unequal variances)
t_stat, p_value = ttest_ind(holiday_sales, non_holiday_sales, equal_var=False)

# Print results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The difference in sales between holiday and non-holiday weeks is statistically significant (p < 0.05).")
else:
    print("The difference in sales between holiday and non-holiday weeks is NOT statistically significant (p >= 0.05).")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load dataset
file_path = "Train_cleaned_processed.csv"  # Update path if needed
df_sales = pd.read_csv(file_path)

# Convert categorical feature Store_Type into numerical
df_sales = pd.get_dummies(df_sales, columns=["Store_Type"], drop_first=True)

# Define final feature selection (dropping Store_Size)
selected_features = ["Lag_52", "Total_MarkDown"] + [col for col in df_sales.columns if "Store_Type" in col]

# Define Features (X) and Target (y)
X = df_sales[selected_features]
y = df_sales["Weekly_Sales"]

# Split into Train (80%) and Test (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Make Predictions
y_pred = model.predict(X_test)

# Evaluate Model Performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Calculate Adjusted RÂ²
n = X_test.shape[0]  # Number of observations
p = X_test.shape[1]  # Number of predictors
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# Print Performance Metrics
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")
print(f"Adjusted RÂ² Score: {adjusted_r2:.4f}")

# Save results
df_results = pd.DataFrame({"Date": df_sales["Date"].iloc[len(y_train):],
                           "Actual_Sales": y_test,
                           "Predicted_Sales": y_pred})
df_results.to_csv("linear_regression_final_results.csv", index=False)

# Plot Actual vs Predicted Sales
plt.figure(figsize=(12, 6))
plt.plot(df_sales["Date"].iloc[len(y_train):], y_test, label="Actual Sales", marker="o")
plt.plot(df_sales["Date"].iloc[len(y_train):], y_pred, label="Predicted Sales", marker="x")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.title("Actual vs Predicted Weekly Sales (Final Model)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Use your existing X, y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 1: Get the effective alphas for pruning
dt = DecisionTreeRegressor(random_state=42)
path = dt.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas[:-1]  # drop last alpha (prunes everything)

# Step 2: Train trees for different alphas
trees = []
for alpha in ccp_alphas:
    model = DecisionTreeRegressor(random_state=42, ccp_alpha=alpha)
    model.fit(X_train, y_train)
    trees.append(model)

# Step 3: Evaluate each tree on test data
test_scores = [mean_squared_error(y_test, tree.predict(X_test)) for tree in trees]
rmse_scores = [np.sqrt(mse) for mse in test_scores]

# Step 4: Plot alpha vs RMSE
plt.figure(figsize=(10, 5))
plt.plot(ccp_alphas, rmse_scores, marker='o', drawstyle="steps-post")
plt.xlabel("ccp_alpha (Pruning Strength)")
plt.ylabel("RMSE on Test Set")
plt.title("Finding Optimal ccp_alpha for Pruning")
plt.grid(True)
plt.show()

# Step 5: Select the best alpha
best_index = np.argmin(rmse_scores)
best_alpha = ccp_alphas[best_index]
print(f"Best ccp_alpha: {best_alpha:.5f}, RMSE: {rmse_scores[best_index]:.2f}")

# Step 6: Final pruned model
pruned_tree = DecisionTreeRegressor(random_state=42, ccp_alpha=best_alpha)
pruned_tree.fit(X_train, y_train)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
from sklearn.metrics import r2_score



# Step 1: Prepare the time series (if not already done)
df_ts = df.groupby('Date')['Weekly_Sales'].sum().reset_index()
df_ts['Date'] = pd.to_datetime(df_ts['Date'])
df_ts.set_index('Date', inplace=True)
y = df_ts['Weekly_Sales'].sort_index()

# Step 2: Split into train and test (e.g., last 12 weeks as test)
train = y[:-12]
test = y[-12:]

# Step 3: Fit the SARIMA model
model = SARIMAX(train,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 52),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Step 4: Forecast
forecast = results.forecast(steps=12)

# Step 5: Evaluate
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

r2 = r2_score(test, forecast)

print("ðŸ“Š SARIMA Model Evaluation:")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")

# Step 6: Plot
plt.figure(figsize=(12, 6))
plt.plot(train, label="Train")
plt.plot(test, label="Actual", color='black')
plt.plot(forecast, label="Forecast", color='orange')
plt.title("SARIMA(1,1,1)(1,1,1,52) Forecast vs Actual")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

file_path = "Train_cleaned_processed.csv"  # Adjust this path if necessary
df = pd.read_csv(file_path)

# Step 1: Prepare the target variable
y = df.groupby('Date')['Weekly_Sales'].sum()
y.index = pd.to_datetime(y.index)
y = y.sort_index()

# Step 2: Prepare exogenous features
exog_features = ['Total_MarkDown', 'IsHoliday_x', 'Store_Size']
exog_df = df.groupby('Date')[exog_features].sum()
exog_df.index = pd.to_datetime(exog_df.index)
exog_df = exog_df.sort_index()
exog_df = exog_df.fillna(0)

# Align with target
exog = exog_df.loc[y.index]

# Step 3: Train-test split (last 12 weeks as test)
y_train, y_test = y[:-12], y[-12:]
exog_train, exog_test = exog[:-12], exog[-12:]

# Step 4: Fit SARIMAX model
model = SARIMAX(y_train,
                exog=exog_train,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 52),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Step 5: Forecast
forecast = results.predict(start=y_test.index[0], end=y_test.index[-1], exog=exog_test)

# Step 6: Evaluation
mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))
r2 = r2_score(y_test, forecast)

print("ðŸ“Š SARIMAX Model with Exogenous Features:")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")

# Step 7: Plot forecast vs actual
plt.figure(figsize=(12, 6))
plt.plot(y_train, label="Train")
plt.plot(y_test, label="Actual", color='black')
plt.plot(forecast, label="Forecast (SARIMAX)", color='orange')
plt.title("SARIMAX Forecast vs Actual (with Exogenous Features)")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.grid(True)
plt.show()
```
